```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
#Use matplotlib to plot the following equation:
#y=x2âˆ’x+2
#You'll need to write the code that generates the x and y points.
#Add an anotation for the point 0, 0, the origin.
x = random.sample(range(0, 21), 10)
x
```

```python
x = np.array(x)
x
```

```python
y = x*2 - x + 2
y
```

```python
plt.plot(x, y, c = 'orange')
plt.xlim(-1)
plt.ylim(-1)
plt.annotate('Origin', xy = (0,0), arrowprops = {'width': 5}, xytext = (8,4))
```

```python
#Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
#y=âˆšx
#y=x3
#y=2x
#y=1/(x+1)
#You can use functions from the math module to help implement some of the equations above.
```

```python
y = x ** .5

plt.scatter(x, y, color = 'rebeccapurple')
plt.title('Square Root of x')
```

```python
r = x * 3

plt.scatter(x, r, color = 'deepskyblue')
plt.title('$x * 3$')
```

```python
m = 2 * x

plt.scatter(x, m, color= 'orange')
plt.title('2x')
```

```python
z = 1 / (1 + x)

plt.scatter(x, z, color = 'magenta')
plt.title('1 / (1 + x)')
```

```python
#Combine the figures you created in the last step into one large figure with 4 subplots.
plt.figure(figsize = (15, 5))

plt.subplot(141)
plt.scatter(x, y, color = 'rebeccapurple')
plt.title('Square Root of x')

plt.subplot(142)
plt.scatter(x, r, color = 'deepskyblue')
plt.title('$x * 3$')

plt.subplot(143)
plt.scatter(x, m, color= 'orange')
plt.title('2x')

plt.subplot(144)
plt.scatter(x, z, color = 'magenta')
plt.title('1 / (1 + x)')
```

```python
#Combine the figures you created in the last step into one figure where each of the 4 equations 
#has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

plt.figure(figsize = (10, 5))

(
    plt.scatter(x, y, color = 'rebeccapurple', label = 'Square Root of x'),
    plt.scatter(x, r, color = 'deepskyblue', label = '$x * 3$'),
    plt.scatter(x, m, color= 'orange', label = '2x'),
    plt.scatter(x, z, color = 'magenta', label = '1 / (1 + x)'),
    plt.plot(x, y, color = 'rebeccapurple'),
    plt.plot(x, r, color = 'deepskyblue'),
    plt.plot(x, m, color= 'orange'),
    plt.plot(x, z, color = 'magenta')
)
plt.title('Math Equations With x!')
plt.legend(loc = 'upper left')
```
