```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.describe()
```

```python
tips.info()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, marker = '*')
plt.title('Tips by Total Bill')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tips ($)')
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', col = 'sex', data=tips, marker = '*')

plt.xlabel('Total Bill ($)')
plt.ylabel('Tips ($)')
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', hue =  'size', data = tips)
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', col = 'time', hue =  'size', style = 'smoker', data = tips)
plt.show()
```

```python
sns.histplot(tips['total_bill'], kde= True)
plt.show()
```

```python
sns.boxplot(data=tips, x='total_bill')
plt.show()
```

```python
sns.boxplot(data= tips, y='tip', x='size')
plt.show()
```

```python
sns.boxplot(data= tips, y='tip', hue = 'smoker', x='size')
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```

```python
sns.jointplot(data= tips, x='total_bill', y='tip')
plt.show()
```
