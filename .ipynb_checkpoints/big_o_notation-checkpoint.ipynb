```python
n = np.array(range(1, 11))

n
```

```python
y = 0*n + 1
z = np.log10(n)
a = n
b = n * np.log10(n)
c = n^2
d = 2 * n
e = scipy.special.factorial(n)
f = n^n
```

```python
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')

(
    plt.plot(n,y, label = "$1$"),
    plt.plot(n,z, label = "$log (n)$"),
    plt.plot(n,a, label = "$n$"),
    plt.plot(n,b, label = "$n * log (n)$"),
    plt.plot(n,c, label = "$n^2$"),
    plt.plot(n,d, label = "$2^n$"),
    plt.plot(n,e, label = "$n!$"),
    plt.plot(n,f, label = "$n^n$"),
)
plt.legend(loc = 'upper left')
```
